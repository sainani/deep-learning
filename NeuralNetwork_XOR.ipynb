##Importing Libraries

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
import numpy as np
```

**Create a Neural network taking the input dataset (A , B) and taking A[XOR]B as the label.**





1. Create x_train (input) as a numpy array taking A and B as inputs.


```python
x_train = np.array([[0,0],[0,1],[1,0],[1,1]])
```

2. Create y_train (label) as a numpy array with A[XOR]B


```python
y_train=np.array([0,1,1,0])
```

3. Create Keras Sequential Model


```python
model=keras.Sequential()
```

4. Create two Dense layers with:
- 4.1 128 Neurons
- 4.2 64 Neurons
- 4.3 Use sigmoid activation function
- 4.4 Provide input shape to this layer (this should be 2 since there are 2 inputs)


```python
model.add(keras.layers.Dense(128,activation='sigmoid',input_shape=(2,)))
model.add(keras.layers.Dense(64,activation='sigmoid',input_shape=(2,)))
```

5. Create output Dense layer with:
- 5.1 Sigmoid activation function
- 5.2 Output has 1 neuron


```python
model.add(keras.layers.Dense(1,activation='sigmoid'))
```

6. Compile the model using:
- 6.1 Loss function as “binary_crossentropy”
- 6.2 Use optimizer as standard gradient descent (SGD)
- 6.3 Use learning rate 0.1
- 6.4 Include metrics as “accuracy”


```python
model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.SGD(learning_rate=0.1),metrics=['accuracy'])
```

7. Fit the model to include the input dataset
- 7.1 Provide the batchsize of 4
- 7.2 Run the model for epochs upto 10000
- 7.3 Use verbose=2 to see the loss and accuracy


```python
model.fit(x_train,y_train,batch_size=4,epochs=10000,verbose=2)
```

8. Finally predict the model using model.predict. 
- 8.2 For prediction use the input dataset  since there is no separate testing data.

```python
print (model.predict(x_train).round())
```

---
### Accuracy

---

```python
loss,accuracy = model.evaluate(x_train,y_train)
```

### Results when using RELU Activation
---

```python
x_train = np.array([[0,0],[0,1],[1,0],[1,1]])
y_train=np.array([0,1,1,0])
model=keras.Sequential()
model.add(keras.layers.Dense(128,activation='relu',input_shape=(2,)))
model.add(keras.layers.Dense(64,activation='relu',input_shape=(2,)))
model.add(keras.layers.Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.SGD(learning_rate=0.1),metrics=['accuracy'])
model.fit(x_train,y_train,batch_size=4,epochs=1000,verbose=2)
```

---
### Prediction
---

```python
print (model.predict(x_train).round())
```

---
### Accuracy
---

```python
loss,accuracy = model.evaluate(x_train,y_train)
```
